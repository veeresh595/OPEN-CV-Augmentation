```python
import numpy as np
import matplotlib.pyplot as plt
import os 
import cv2
DATADIR='C:\\Users\\Lenovo\\OneDrive\\Desktop\\Capstone project\\dataset'
CATEGORIES=['Bacterial leaf blight','Brown spot','Leaf smut']
```

```python
os.getcwd()
```

```python
img_array=[]
for categories in CATEGORIES:
    path=os.path.join(DATADIR,categories)
    for img in os.listdir(path):
        img_array=cv2.imread(os.path.join(path,img),cv2.IMREAD_GRAYSCALE)
        print(img_array)
        plt.imshow(img_array,cmap ='gray')
        plt.show()
        break
    break   
```

```python
image_size=np.arange(55,255,10)
def resize_image():
        for i in range(len(image_size)):
            new_array=cv2.resize(img_array,(image_size[i],image_size[i]))
            print(new_array)
            plt.imshow(new_array,cmap ='gray')
            plt.xlabel('pixels')
            plt.ylabel(image_size[i])
            plt.legend('pixels')
            plt.show()
resize_image()        
```

```python
flip_style=[-1,0,1]
def flip_image():
    for i in range(len(flip_style)):
        flip_image=cv2.flip(img_array,flip_style[i])
        print(flip_image)
        plt.imshow(flip_image,cmap ='gray')
        plt.xlabel('pixels')
        plt.show()
flip_image()        
```

```python
loop=[1]
def rotation_image():
    for i in range(len(loop)):
        img_array=cv2.imread(os.path.join(path,img),cv2.IMREAD_GRAYSCALE)
        height,width=img_array.shape[:2]
        rotation_matrix=cv2.getRotationMatrix2D((width/2,height/2),45,.5)
        rotated_image=cv2.warpAffine(img_array,rotation_matrix,(width,height))
        print(rotated_image)
        plt.imshow(rotated_image)
        plt.show()
rotation_image()        
```

```python
loop=[1]
def rotation_image():
    for i in range(len(loop)):
        img_array=cv2.imread(os.path.join(path,img),cv2.IMREAD_GRAYSCALE)
        sharpening=np.array([[-1,-1,-1],
                             [-1,10,-1],
                             [-1,-1,-1] ])
        sharpened_image=cv2.filter2D(img_array,-1,sharpening)
        print(sharpened_image)
        plt.imshow(sharpened_image)
        plt.show()
rotation_image()  
```

```python
training_data=[]
image_size=255
def create_training_data():
    for categories in CATEGORIES:
        path=os.path.join(DATADIR,categories)
        class_index=CATEGORIES.index(categories)
        for img in os.listdir(path):
            img_array=cv2.imread(os.path.join(path,img),cv2.IMREAD_GRAYSCALE)
            new_array=cv2.resize(img_array,(image_size,image_size))
            training_data.append([new_array,class_index])
            
create_training_data()        
        
        
        
```
